Import Libraries

```python
import pandas as pd
import sqlite3
```

Load the file

```python
df = pd.read_excel("Cleaned_Dataset.xlsx")

df.head()
```

SQL Database

```python
conn = sqlite3.connect("ecommerce.db")
```


Convert Excel Data into SQL Table

```python
df.to_sql("orders", conn, if_exists="replace", index=False)
```

SQL QUERIES

Query 1 - SELECT *

```python
query = """
SELECT *
FROM orders
LIMIT 5
"""

pd.read_sql(query, conn)
```

Query 2 - SELECT Specific Columns

```python
query = """
SELECT Product, TotalPrice
FROM orders
LIMIT 10
"""

pd.read_sql(query, conn)
```

Query 3 - WHERE

```python
query = """
SELECT *
FROM orders
WHERE PaymentMethod='Online'
LIMIT 10
"""

pd.read_sql(query, conn)
```

Query 4 - WHERE Product

```python
query = """
SELECT *
FROM orders
WHERE Product='Printer'
LIMIT 10
"""

pd.read_sql(query, conn)
```

Query 5 - ORDER BY

```python
query = """
SELECT OrderID, Product, TotalPrice
FROM orders
ORDER BY TotalPrice DESC
LIMIT 10
"""

pd.read_sql(query, conn)
```

Query 6 - COUNT

```python
query = """
SELECT COUNT(*) AS TotalOrders
FROM orders
"""

pd.read_sql(query, conn)
```

Query 7 - AVG

```python
query = """
SELECT AVG(TotalPrice) AS AverageOrderValue
FROM orders
"""

pd.read_sql(query, conn)
```

Query 8 - SUM

```python
query = """
SELECT SUM(TotalPrice) AS TotalRevenue
FROM orders
"""

pd.read_sql(query, conn)
```

Query 9 - GROUP BY Product

```python
query = """
SELECT Product,
COUNT(*) AS TotalOrders
FROM orders
GROUP BY Product
ORDER BY TotalOrders DESC
"""

pd.read_sql(query, conn)
```

Query 11 - GROUP BY Referral Source

```python
query = """
SELECT ReferralSource,
COUNT(*) AS Customers
FROM orders
GROUP BY ReferralSource
ORDER BY Customers DESC
"""

pd.read_sql(query, conn)
```

# Conclusion

SQL queries such as SELECT, WHERE, ORDER BY, GROUP BY, COUNT, SUM, and AVG were successfully used to analyze the e-commerce dataset. The analysis helped identify product demand, payment preferences, referral source performance, and overall business metrics.
